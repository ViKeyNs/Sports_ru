```python
group = data.groupby('Type')
groupType = group.agg([np.mean, np.max, np.min, np.median, mode])
```

```python
def mode(numbers):
    numbers = list(numbers)
    counts = {k:numbers.count(k) for k in set(numbers)}
    modes = sorted(dict(filter(lambda x: x[1] == max(counts.values()), counts.items())).keys())
    return modes
```

```python
whole = data.describe().loc[['mean','max','min','50%']]
a = pd.DataFrame(np.array([mode(data[i]) for i in data.columns]),columns=['mode']).transpose()
a.columns = data.columns
a.drop('Type',axis=1)
wholewmode = whole.append(a)
wholewmode = wholewmode.drop('Type',axis=1)
```

```python
def order(listik):
    orderlist = sorted(list(listik),reverse=True)
    spis = [orderlist.index(i) for i in list(listik)]
    return spis
```
